```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import numpy as np
from concurrent.futures import ThreadPoolExecutor
import threading
def generate_fingerprints(smiles_list, radius=2, n_bits=2048, n_jobs=-1):
    """
    Generate Morgan fingerprints for a list of SMILES strings using multithreading.
    
    Parameters:
    -----------
    smiles_list : list or array-like
        List of SMILES strings
    radius : int, default=2
        Morgan fingerprint radius
    n_bits : int, default=2048
        Number of bits in the fingerprint
    n_jobs : int, default=-1
        Number of parallel jobs. If -1, use all available cores.
        
    Returns:
    --------
    list
        List of numpy arrays containing fingerprints (or None for invalid SMILES)
    """
    # Determine number of threads
    if n_jobs == -1:
        import multiprocessing
        n_jobs = multiprocessing.cpu_count()
    
    # Function to process a single SMILES string
    def process_smiles(smiles):
        try:
            mol = Chem.MolFromSmiles(smiles)
            if mol is None:
                return None
            
            # Generate fingerprint
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=n_bits)
            
            # Convert to numpy array
            arr = np.zeros((n_bits,), dtype=np.int8)
            DataStructs.ConvertToNumpyArray(fp, arr)
            
            return arr
        except Exception as e:
            print(f"Error processing SMILES {smiles}: {e}")
            return None
    
    # Process SMILES in parallel
    results = []
    with ThreadPoolExecutor(max_workers=n_jobs) as executor:
        results = list(executor.map(process_smiles, smiles_list))
    
    return results
```

```python
import requests
from urllib.parse import quote

def smiles_to_name(smiles):
    escaped_smiles = quote(smiles)
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/smiles/{escaped_smiles}/property/IUPACName/JSON"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if "PropertyTable" in data:
            return data["PropertyTable"]["Properties"][0]["IUPACName"]
    return None
```

```python
print(smiles_to_name('C1=CC=CC(=C1C(O)=O)OC(C)=O')) 
```

```python
import pubchempy as pcp

def name_to_smiles(drug_name):
    compounds = pcp.get_compounds(drug_name, 'name')
    if compounds:
        return compounds[0].canonical_smiles
    return None

# Example usage
print(name_to_smiles('Aspirin'))  # Output: CC(=O)OC1=CC=CC=C1C(=O)O
```
